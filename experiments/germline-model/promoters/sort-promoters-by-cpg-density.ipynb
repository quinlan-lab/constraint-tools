```python
import pandas as pd 

CONSTRAINT_TOOLS = '/scratch/ucgd/lustre-work/quinlan/u6018199/constraint-tools'
promoters_grch37_filename = f'{CONSTRAINT_TOOLS}/download-process-data/promoters/promoters.grch37.csv'
promoters_grch37 = pd.read_csv(promoters_grch37_filename, sep=',')

def compute_region_grch37(row):
    return f'{row["chr"]}:{int(row["start"])+1}-{row["end"]}'
    
promoters_grch37['region_grch37'] = promoters_grch37.apply(compute_region_grch37, axis=1)
promoters_grch37 = promoters_grch37.drop(columns=['chr', 'start', 'end'])
promoters_grch37
```

```python
promoters_grch37[promoters_grch37['region_grch37'] == 'chr15:42118284-42122282'] 
```

```python
promoters_coordinates_grch38_grch37_filename = f'{CONSTRAINT_TOOLS}/download-process-data/promoters/promoters.coordinates.grch38.grch37.bed'
promoters_coordinates_grch38_grch37 = pd.read_csv(promoters_coordinates_grch38_grch37_filename, sep='\t', header=None)
promoters_coordinates_grch38_grch37 = promoters_coordinates_grch38_grch37.drop(columns=[4])
promoters_coordinates_grch38_grch37.columns = ['chromosome_grch38', 'start_grch38', 'end_grch38', 'region_grch37']
promoters_coordinates_grch38_grch37
```

```python
def compute_region_grch38(row):
    return f'{row["chromosome_grch38"]}:{int(row["start_grch38"])+1}-{row["end_grch38"]}'
    
def compute_region_grch38_clipped(row):
    flank = 1500
    return f'{row["chromosome_grch38"]}:{int(row["start_grch38"])+1+flank}-{int(row["end_grch38"])-flank}'

def compute_region_grch38_length(row):
    return int(row["end_grch38"]) - int(row["start_grch38"])

promoters_coordinates_grch38_grch37['region_grch38'] = promoters_coordinates_grch38_grch37.apply(compute_region_grch38, axis=1)
promoters_coordinates_grch38_grch37['region_grch38_length'] = promoters_coordinates_grch38_grch37.apply(compute_region_grch38_length, axis=1)
promoters_coordinates_grch38_grch37['region_grch38_clipped'] = promoters_coordinates_grch38_grch37.apply(compute_region_grch38_clipped, axis=1)
promoters_coordinates_grch38_grch37 = promoters_coordinates_grch38_grch37.drop(columns=['chromosome_grch38', 'start_grch38', 'end_grch38'])
promoters_coordinates_grch38_grch37
```

```python
promoters_coordinates_grch38_grch37[promoters_coordinates_grch38_grch37['region_grch37'] == 'chr15:42118284-42122282'] 
```

```python
promoters_coordinates_grch38_grch37 = promoters_coordinates_grch38_grch37.drop_duplicates()
promoters_coordinates_grch38_grch37
```

```python
promoters_grch38_grch37 = pd.merge(
  promoters_grch37, 
  promoters_coordinates_grch38_grch37, 
  on='region_grch37', 
  how='inner'
)
promoters_grch38_grch37
```

```python
promoters_grch38 = promoters_grch38_grch37.drop(columns=['region_grch37'])
promoters_grch38
```
